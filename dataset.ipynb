```python
import pandas as pd
# 数据导入
df=pd.read_csv('new.csv',encoding='utf-8')
# 查看数据类型
df.info()
#查看空行
print(pd.isnull(df).any())
#如果检测的位置为空，则返回Ture
print(df.isnull())
#删除空行all是全部，any任意axis : 删除行还是列，{0 or ‘index’, 1 or ‘columns’}, default 0
df1=df.dropna(axis=0, how='any')
#df.head(50)
#查看空行
print(pd.isnull(df).any())
print(len(df))
```

```python
import random
df1.head(5)
#删除cast列
df1.drop(['cast'],axis=1,inplace=True)
#重新创建索引
df1=df1.reset_index(drop=True)
print(len(df1))
for i in range(len(df1)):
    #print(i)
    df1.loc[i,('cast1')]=random.randint(1,1000) 
    df1.loc[i,['cast2']]=random.randint(1,1000) 
    df1.loc[i,['cast3']]=random.randint(1,1000) 
  
#df1.to_csv('11.csv')
```

```python
df1.head(5)
df2=df1.groupby('country')
print(len(df2))
print(type(df2))
df3=pd.DataFrame(df2,columns=('a','b'))
df3
```

```python
#print(df3['a'])
guojia={}
a=0
for i in df3['a']:
    guojia[i]=a
    a+=1
#print(guojia)
for i in range(len(df1)):
    #print(i)
    for k,v in guojia.items():
        if k== df1.loc[i,('country')]:
            df1.loc[i,('country1')]=v
df1.head(100)
```

```python
df1['duration1']=df1['duration'].str.split(" ",expand=False).str[0]
df1.head(10)
```

```python
df1['listed_in1']=df1['listed_in'].str.split(",",expand=False).str[0]
df1.head(10)
```

```python
df2=df1.groupby('listed_in1')
print(len(df2))
print(type(df2))
df3=pd.DataFrame(df2,columns=('a','b'))
df3
```

```python
#print(df3['a'])
guojia={}
a=0
for i in df3['a']:
    guojia[i]=a
    a+=1
#print(guojia)
for i in range(len(df1)):
    #print(i)
    for k,v in guojia.items():
        if k== df1.loc[i,('listed_in1')]:
            df1.loc[i,('listed_in2')]=v
df1.head(100)
```

```python
for i in range(len(df1)):
    #print(i)
    if df1.loc[i,('listed_in2')]>=0 and df1.loc[i,('listed_in2')]<=15 and df1.loc[i,('score')]<5:
            df1.loc[i,('score1')]=int(df1.loc[i,('score')])+1
    else:
        df1.loc[i,('score1')]=df1.loc[i,('score')]
df1.to_csv('new.csv')
```
